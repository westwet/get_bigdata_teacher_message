```python
from selenium import webdriver
import time
import bs4
import pandas as pd
from bs4 import BeautifulSoup
import requests
soup = requests.get('http://www.cs.sjtu.edu.cn/Faculty.aspx')
soup.encoding = 'utf-8'
te = BeautifulSoup(soup.text)
```

```python
te
```

```python
l = te.find_all('a',{'href':True})
```

```python
l = [i.get('href') for i in l]
```

```python
url_all = [i for i in l[104:122]]
```

```python
teacher_all_message = []
for teacher_url in  url_all:
    page = requests.get('http://www.cs.sjtu.edu.cn/'+teacher_url)
    page.encoding = 'utf-8'
    te = BeautifulSoup(page.text)
    l = [i.text for i in te.find('div',{'class':'w480 fr lh300 f14'}).find_all('p')]
    l.append(te.find('div',{'id':'con_one_1'}).text)
    teacher_all_message.append(l)
```

```python
pd.DataFrame(teacher_all_message).to_excel('C:\\Users\\NYDSJ\\Desktop\\上交老师并行与分布式计算.xls')
```

```python
teacher_all_message
```

```python
new_teacher = []
for i in teacher_all_message:
    l = []
    for j in i:
        k = j.replace('      ','')
        k = k.replace('\r','')
        k = k.replace('\n','')
        k = k.replace('\t','')
        l.append(k)
    new_teacher.append(l)
```
